```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfs = []
for body in ["biped-5x4", "block-5x5", "tripod-5x5", "worm-5x2", "evo-body-5x5"]:
    for controller in ["global", "local", "local2"]:
        for seed in range(10):
            tmp_df = pd.read_csv(f"../results/{body}_{controller}_{seed}.csv")
            tmp_df["body"] = body
            tmp_df["seed"] = seed
            tmp_df["controller"] = controller
            dfs.append(tmp_df)
df = pd.concat(dfs)
df.head()
```

```python
sns.relplot(
    data=df[df["body"] != "evo-body-5x5"],
    x="iteration",
    y="max_fitness",
    hue="controller",
    col="body",
    col_wrap=2,
    kind="line",
    estimator="median",
    palette="Set1",
    errorbar=("pi", 50),
)
plt.show()
```

```python
sns.lineplot(
    data=df[df["body"] == "evo-body-5x5"],
    x="iteration",
    y="max_fitness",
    hue="controller",
    estimator="median",
    errorbar=("pi", 50),
    palette="Set1"
)
plt.show()
```
