```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfs = []
for body in ["biped-3x2", "biped-4x3", "worm-5x2"]:
    for controller in ["global", "local"]:
        for seed in range(7):
            tmp_df = pd.read_csv(f"../results/{body}_{controller}_{seed}.csv")
            tmp_df["body"] = body
            tmp_df["seed"] = seed
            tmp_df["controller"] = controller
            dfs.append(tmp_df)
df = pd.concat(dfs)
df.head()
```

```python
g = sns.relplot(
    data=df,
    x="iteration",
    y="max_fitness",
    hue="controller",
    col="body",
    kind="line",
    estimator="median",
    palette="Set1"
)
plt.show()
```
